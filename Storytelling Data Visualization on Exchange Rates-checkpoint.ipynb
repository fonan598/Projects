# Storytelling Data Visualization on Exchange Rates

- The dataset we'll use describes Euro daily exchange rates between 1999 and 2021. The euro (symbolized with €) is the official currency in most of the countries of the European Union.
- If the exchange rate of the euro to the US dollar is 1.5, you get 1.5 US dollars if you pay 1.0 euro (one euro has more value than one US dollar at this exchange rate).
- Daria Chemkaeva put together the data set and made it available on Kaggle — the data source is the European Central Bank. Note that the dataset gets regular updates — we downloaded it on September 2024.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.style as style
%matplotlib inline
```

```python
exchange_rates = pd.read_csv('euro-daily-hist_1999_2022.csv')
```

```python
print(exchange_rates.info())
print(exchange_rates.tail())
```

- we have 40 different currencies against the euro
- we have 6655 days (observations or rows)
- 18 currencies have null values in them
- only 3 of the columns are numeric...the rest are objects

- For this project we will focuse on using the USD column
- So we will clean only the Date column and the USD column

```python
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar',
                               'Period\\Unit:': 'Time'},
                      inplace=True) #renaming the columns
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time']) #change to datetime
exchange_rates.sort_values('Time', inplace=True) #sort values in ascending order by timne
exchange_rates.reset_index(drop=True, inplace=True)
```

```python
exchange_rates
```

```python
#isolate just the time and US Dollar columns
euro_to_dollar = exchange_rates.iloc[:, [0, 39]]
euro_to_dollar.info()
```

```python
euro_to_dollar['US_dollar'].value_counts()
#from the above line, we see there is 62 values with '-'

#so I will remove those
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-']
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)
```

```python
euro_to_dollar.info() #nice and clean
```

```python
#now going to calculate the monthly rolling mean for our data
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(60).mean()


fig, ax = plt.subplots(figsize=(15,8))
ax.plot(euro_to_dollar['Time'], euro_to_dollar['rolling_mean'])
plt.show()
```

- I am going to explore the rate before and leading up to the 2008 financial crisis
- mainly using the data from 1999-2009
- 10 year span of data

```python
eurotodollar_sub = euro_to_dollar.iloc[:, 0:2500]
eurotodollar_sub
```

```python
fig, axes = plt.subplots(nrows = 4, ncols=1, figsize=(8,15))

for ax in axes:
    ax.plot(euro_to_dollar['Time'], 
            euro_to_dollar['rolling_mean'],
           linewidth = 1)
    ax.axis(xmin=euro_to_dollar['Time'][0], xmax=euro_to_dollar['Time'][2500])
    #ax.set_xticklabels([])
    ax.set_yticklabels([])
    ax.tick_params(bottom=False, left=False)
    for location in ['left', 'right', 'top', 'bottom']:
        ax.spines[location].set_visible(False)
    
#axes[0].text(x=0, y=-0.5, s='1.18',size=100)

plt.show()
```
